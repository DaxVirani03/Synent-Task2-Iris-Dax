# Dax Virani | Iris Dataset Visual Analytics
A polished notebook for species comparison, feature distribution, correlation structure, and dimensionality reduction.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

sns.set_theme(style="whitegrid")
plt.rcParams.update({
    "figure.facecolor": "#F8FAFC",
    "axes.facecolor": "#FFFFFF",
    "font.family": "DejaVu Sans",
    "axes.titleweight": "bold"
})

iris = load_iris(as_frame=True)
df = iris.frame.copy()
df["species"] = df["target"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
df = df.drop(columns="target").rename(columns={
    "sepal length (cm)": "sepal_length",
    "sepal width (cm)": "sepal_width",
    "petal length (cm)": "petal_length",
    "petal width (cm)": "petal_width",
})
df.head()
```

## Dataset Profile
The Iris data is perfectly balanced, which makes it ideal for clear visual comparison and prototype modeling.

```python
profile = pd.DataFrame({"rows": [len(df)], "columns": [df.shape[1]], "missing": [int(df.isna().sum().sum())]}, index=["iris"])
display(profile)
display(df.groupby("species").mean(numeric_only=True).round(2))
display(df.describe().round(2))
```

## Advanced Visuals
Pairwise separation, correlation structure, and a PCA projection show how strongly the petal features drive class separation.

```python
palette = {"setosa": "#2563EB", "versicolor": "#16A34A", "virginica": "#DC2626"}
sns.pairplot(df, hue="species", palette=palette, corner=True, plot_kws={"alpha": 0.8, "s": 45})
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.heatmap(df.drop(columns="species").corr(), annot=True, cmap="RdYlGn", vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title("Feature Correlation")

scaled = StandardScaler().fit_transform(df.drop(columns="species"))
pca = PCA(n_components=2, random_state=42)
pc = pca.fit_transform(scaled)
pca_df = pd.DataFrame(pc, columns=["PC1", "PC2"]).join(df["species"])
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="species", palette=palette, ax=axes[1], s=70)
axes[1].set_title(f"PCA Projection ({pca.explained_variance_ratio_.sum():.1%} variance)")
plt.tight_layout()
plt.show()
```

## Closing Summary
Setosa is isolated by petal measurements, while versicolor and virginica overlap more and benefit from multivariate models.
